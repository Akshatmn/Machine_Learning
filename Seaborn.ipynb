```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.set_theme(style = "white")
sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip"
)

plt.title("Tip")
```

```python
sns.set_theme(style = "dark")
sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip"
)

plt.title("Tip")
```

```python
sns.set_theme(style = "ticks")
sns.scatterplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    # hue = "smoker",
    # hue = "day",
    # hue = "size", 
    hue = "time"
    # pallete = {"Yes" : "red", "No" : "blue"}
)

plt.title("Tip Vs Bill Graph")
```

```python
sns.set_theme()
sns.lineplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    hue = "smoker",
    # hue = "day",
    # hue = "size", 
    # hue = "time"
    # pallete = {"Yes" : "red", "No" : "blue"}
)

plt.title("Tip Vs Bill Graph")
```
