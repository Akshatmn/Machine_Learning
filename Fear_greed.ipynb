```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_classification["classification"] = le.fit_transform(df_classification["classification"])
```

```python
df_classification.isnull().sum()
```

```python
df_classification["classification"].value_counts().plot(
    kind="bar"
)
plt.title("No of trades based on Market Sentiment")
```

```python
df_hist_data.groupby("Side")["Closed PnL"].mean().plot(kind="bar")
plt.title("Average Closed PnL by Trade Side")
plt.xlabel("Trade Side")
plt.ylabel("Average Closed PnL")
plt.show()
```

So according to me, closed PnL of sell side is very high as compared to buy side. Hence when trades less time will make better profit rather than of taking longer time
